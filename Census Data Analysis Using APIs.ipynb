```python
import pandas as pd
import numpy as np
import json
import requests
from time import sleep
from statistics import mean


msas = pd.read_csv('msas.csv')
```

```python
tableids = ['B01001','B19001', 'B03002']
```

```python
print('Find the Specific Tables You Want ')
print('Itterate through each geoids and calculate the sum for each city for each parameter')

listofmsas = msas['CBSA'].tolist()
listB01001 = []
for anmsa in listofmsas:
    url = 'http://api.censusreporter.org/1.0/data/show/latest?table_ids=%s&geo_ids=31000US%s' % ('B01001', anmsa)
    print(url)
    censusreporter = requests.get(url)
    loadedjson = json.loads(censusreporter.content)
    total = 0
    total += loadedjson['data']["31000US%s" % (anmsa)]['B01001']['estimate']['B01001007']
    total += loadedjson['data']["31000US%s" % (anmsa)]['B01001']['estimate']['B01001008']
    total += loadedjson['data']["31000US%s" % (anmsa)]['B01001']['estimate']['B01001009']
    total += loadedjson['data']["31000US%s" % (anmsa)]['B01001']['estimate']['B01001010']
    total += loadedjson['data']["31000US%s" % (anmsa)]['B01001']['estimate']['B01001011']
    total += loadedjson['data']["31000US%s" % (anmsa)]['B01001']['estimate']['B01001012']
    total += loadedjson['data']["31000US%s" % (anmsa)]['B01001']['estimate']['B01001031']
    total += loadedjson['data']["31000US%s" % (anmsa)]['B01001']['estimate']['B01001032']
    total += loadedjson['data']["31000US%s" % (anmsa)]['B01001']['estimate']['B01001033']
    total += loadedjson['data']["31000US%s" % (anmsa)]['B01001']['estimate']['B01001034']
    total += loadedjson['data']["31000US%s" % (anmsa)]['B01001']['estimate']['B01001035']
    total += loadedjson['data']["31000US%s" % (anmsa)]['B01001']['estimate']['B01001036']
    listB01001.append(total)
    sleep(.5)
    
listB19001 = []
```

```python
for anmsa in listofmsas:
    url = 'http://api.censusreporter.org/1.0/data/show/latest?table_ids=%s&geo_ids=31000US%s' % ('B19001', anmsa)
    print(url)
    censusreporter = requests.get(url)
    loadedjson = json.loads(censusreporter.content)
    total = 0
    total += loadedjson['data']["31000US%s" % (anmsa)]['B19001']['estimate']['B19001011']
    total += loadedjson['data']["31000US%s" % (anmsa)]['B19001']['estimate']['B19001012']
    listB19001.append(total)
    sleep(.5)
```

```python
listB03002 = []

for anmsa in listofmsas:
    url = 'http://api.censusreporter.org/1.0/data/show/latest?table_ids=%s&geo_ids=31000US%s' % ('B03002', anmsa)
    print(url)
    censusreporter = requests.get(url)
    loadedjson = json.loads(censusreporter.content)
    total = 0
    total += loadedjson['data']["31000US%s" % (anmsa)]['B03002']['estimate']['B03002011']
    listB03002.append(total)
    sleep(.5)
```

```python
msas['B01001'] = listB01001
msas['B19001'] = listB19001
msas['B03002'] = listB03002
```

```python
print('====== Question 1 =======')
print(' ======== Parameter 1 - HH Income ======')
B19001Max = msas['B19001'].max()
msas[msas['B19001'] == B19001Max]
print('New York, New Jersey, Newark, Pennsylvania area')
print(B19001Max)
```

```python
print('========= Question 2 ========')
print('========== Parameter 2 - Age ========')
B01001Max = msas['B01001'].max()
msas[msas['B01001'] == B01001Max]
print('New York, New Jersey, Newark, Pennsylvania area')
print(B01001Max)
```

```python
print(' ========= Question 3 =========')
print('========= Parameter 4 - Hispanic Population ========')
B03002Max = msas['B03002'].max()
msas[msas['B03002'] == B03002Max]
print('Los Angeles Area') 
print(B03002Max)
```

```python
print('===========  Question 4 ==========')
print('========== Calculate Avg Percentages of 3 Parameters========')
totalB01001 = sum(listB01001)
msas['B01001Perc'] = (msas['B01001'] / totalB01001) * 100

totalB19001 = sum(listB19001)
msas['B19001Perc'] = (msas['B19001'] / totalB19001) * 100

totalB03002 = sum(listB03002)
msas['B03002Perc'] = (msas['B03002'] / totalB03002) * 100


msas['avgPar'] = (msas['B01001Perc'] + msas['B19001Perc'] + msas['B03002Perc']) / 3

avgParMax = msas['avgPar'].max()
msas[msas['avgPar'] == avgParMax]
print('New York, New Jersey, Newark, Pennsylvania area')
print(avgParMax)
```
